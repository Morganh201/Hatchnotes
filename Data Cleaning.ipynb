# Data Cleaning


Data Cleaning is one of the most important aspects in analysing data, as better quality data can lead to better and more accurate data that can be used to more accurately create predictions and forecasts.

```python
import pandas as pd
import numpy as np
```

If we want to only focus on a select few variables for our analysis, we can drop columns from the DataFrame using the drop() function

```python
# We import the csv file and convert it to a DataFrame

df = pd.read_csv("BL-Flickr-Images-Book.csv")
to_drop = ['Edition Statement',
           'Corporate Author', 
           'Corporate Contributors', 
           'Former owner', 
           'Engraver', 
           'Contributors', 
           'Issuance type', 
           'Shelfmarks']
df.drop(to_drop, axis=1, inplace=True)
df
```

We can change the index of the dataset, so instead of the indexes to go from 0 to 8286, we can change them to their identifier numbers on the first column.

```python
df.set_index('Identifier', inplace=True)
df
```

Now that we have removed columns that we do not want, we can now clean specific columns so that the computer can get a better understanding of the dataset. We can first look at the columns where a numerical value can be placed on columns which have the object data type.

```python
df.loc[:, 'Date of Publication']
```

The loc[] function is labeled as [start:stop, 'column_name'], where it returns the rows from index start to index stop and the column values. As shown, because this column represents the year, it makes sense to convert this data from the data type object to the data type int. But first we must deal with the different years that a row may have, for example, 1879 [1878], or 1831, 32. To convert a string, we can use regular expressions: https://docs.python.org/3/howto/regex.html

```python
regex = r'^(\d{4})' # meaning we take the first four digits of any string
temp = df['Date of Publication'].str.extract(regex, expand=False)
df['Date of Publication'] = pd.to_numeric(temp)
df['Date of Publication'].head()
```

From above, we first defined a regular expression that takes the first four digits of a string, then we applied that regular expression to all values of the Series Date of Publication. Thus, changing the value to a single year, which is what we want. Because the values are still object types, through pandas, we changed the data type to a float type, and now this column is ready to be used for analysis.

Another problem in data is when some of the rows in the data is missing, and has null values. It would be easy to simply remove those rows, however this can result in the removal of important information. A better way to fill these null values would be to replace them with the mean value.

```python
df.info()

# It is not wise to fill in these null values with the mean in this example
df['Date of Publication'] = df['Date of Publication'].fillna(df['Date of Publication'].mean())
```

Before we enter the data into analysis, we should remove obvious outliers from continuous data. Because most data follows the normal distribution, we can remove data that is above or below 3 standard deviations from the mean.

```python
from scipy import stats
from matplotlib import pyplot as plt

def remove_outliers(column_name, df_in):
    df_in[f'{column_name}_zscore'] = np.abs(stats.zscore(df_in[f'{column_name}']))
    df_clean = df_in[df_in[f'{column_name}_zscore']<3]
    df_clean = df_clean.drop(f'{column_name}_zscore', axis=1)
    return df_clean

df = remove_outliers('Date of Publication', df)
plt.hist(df['Date of Publication'], density=True)
plt.grid('True')
plt.xlabel("Date of Publication")
plt.ylabel("Probability")
plt.show()
```

Because machines can only interpret numbers, it is important that we convert object types into numerical values so that the computer will be able to make forecasts. One such technique is to implement one hot encoding, where it creates new variables that take on values 0 and 1 to represent the categorical values from the original dataset. If there are 3 values to a feature, then one hot encoding will create 3 new features, each corresponding to a value of the feature, and will take on either 0 or 1, depending on whether or not the value is present or not.

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore')# ignores all values with nan values

def onehot(encoder, df, column_name):
    encoder_df = pd.DataFrame(encoder.fit_transform(df[[column_name]]).toarray())
    encoder_df = encoder_df.add_prefix(f'{column_name}_')
    df = df.join(encoder_df)
    df = df.drop(column_name)
    return df

# Another technique is the Label Encoder, where we simply give a number to every label in the category. Note that the computer
# will automatically give more value to a larger number, thus this technique should only be used with ordinal categories

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
def label(encoder, df, column_name):
    df[f'{column_name}'] = encoder.fit_transform(df[f'{column_name}'])
    return df
```

However, there are many variables, especially in this dataset, where there are many different values for a feature, and thus one hot encoding would increase drastically the dimensionality of the dataset. Therefore, we must find an encoding technique that will reduce the dimensionality of the dataset relative to the number of values in a feature. A good technique for this is feature hashing, where we convert a string of characters to a unique hash value through a hash function.

```python
from sklearn.feature_extraction import FeatureHasher
h = FeatureHasher(n_features=8, input_type='string')

hashed_feature = h.transform([df['Place of Publication']])
hashed_feature = hashed_feature.toarray()
df = pd.concat([df, pd.DataFrame(hashed_feature)], axis=1)
df
```
